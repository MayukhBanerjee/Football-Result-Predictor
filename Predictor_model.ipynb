```python
#ML can't work with object; hence now we know that we need to clean the data
df["date"] = pd.to_datetime(df["date"])
```

```python
df["venue_code"] = df["venue"].astype("category").cat.codes
df["opp_code"] = df["opponent"].astype("category").cat.codes
df["hour_code"] = df["time"].str.replace(":.+", "", regex = True).astype("int") #regex function to not consider minutes in the time
df["day_code"] = df["date"].dt.dayofweek
```

```python
#SETTING UP TARGET 
df["target"] = (df["result"] == "W").astype("int")
df
```

```python
# for we use RF because it can pick up on a non - linear dataframe better than one with a linear one
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 50 , min_samples_split = 10 , random_state = 1)

#anything before 2022
train = df[df["date"]< "2022-01-01"]

#anything in 2022
test = df[df["date"] > "2022-01-01"]

#list of the predictor columns that we created

predictors = ["venue_code", "opp_code", "hour_code", "day_code"]
```

```python
#fit our random forest
rf.fit(train[predictors] , train["target"])
```

```python
preds = rf.predict(test[predictors])
```

```python
#how do you determine the accuracy of the model?
from sklearn.metrics import accuracy_score
acc = accuracy_score (test["target"], preds)
acc
```

```python
#implementing rolling avverages
grouped_df = df.groupby("team")
group = grouped_df.get_group("Liverpool").sort_values("date")
```

```python
def avg(group, cols, new_cols):
    group = group.sort_values("date")
    stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]

avg(group, cols, new_cols)
```
